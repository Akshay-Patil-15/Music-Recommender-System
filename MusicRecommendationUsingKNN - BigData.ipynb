```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
p = 1
df_main = pd.read_csv('spotify_dataset.csv', error_bad_lines=False, warn_bad_lines=False, skiprows=lambda i: i>0 and random.random() > p)
df_main.head()
```

```python
df_main.columns = df_main.columns.str.replace('"', '')
df_main.columns = df_main.columns.str.replace('name', '')
df_main.columns = df_main.columns.str.replace(' ', '')
df_main.columns
```

```python
df_main["TrackArtist"] = df_main["track"] + " - " + df_main["artist"]
df_main
```

```python
len(df_main['TrackArtist'].unique()) # Unique users
```

```python
df_main['TrackArtist'].value_counts() # Popular tracks
```

```python
len(df_main['user_id'].unique()) # Unique users
```

```python
df_main['frequency'] = df_main['TrackArtist'].map(df_main['TrackArtist'].value_counts())
df_main
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
df_main[["frequency"]] = min_max_scaler.fit_transform(df_main[["frequency"]])
df_main
```

```python
import hashlib
df_main['track_id'] = [hashlib.blake2b(str(track).encode('utf-8')).hexdigest()[:24] for track in df_main["TrackArtist"]]
df_main
```

```python
df_main.to_csv('SpotifyRefined.csv', index = False)
```

```python
import pandas as pd
import random
#df_main = pd.read_csv('SpotifyRefined.csv')
df_main = pd.read_csv('SpotifyRefined.csv', skiprows=lambda i: i>0 and random.random() > 0.01)
df_main
```

```python
len(df_main['track_id'].unique()) # Unique users
```

```python
len(df_main['user_id'].unique()) # Unique users
```

```python
len(df_main['TrackArtist']) # Unique users
```

```python
df_main = df_main.drop_duplicates(['TrackArtist'])
len(df_main['TrackArtist']) # Unique users
```

```python
df_main
```

```python
df_main.dropna()
df_main
```

```python
df_pivot = df_main.pivot(index='user_id',columns='track_id',values='frequency')
df_pivot
```

```python
df_pivot = df_pivot.fillna(0)
df_pivot
```

```python
from sklearn.neighbors import NearestNeighbors
knn=NearestNeighbors(n_neighbors=5,metric='cosine')
Model=knn.fit(df_pivot)
```

```python
df_pivot[df_pivot.index == 1]
```

```python
user = df_pivot.iloc[1,]
```

```python
distances, indices = Model.kneighbors([user])
distances, indices
```

```python
user
```

```python
neighbors=[]
for item in indices[0][1:]:
    neighbors.append(df_main[df_main.index == item].user_id.values[0])
neighbors
```

```python
neighbors
```

```python
df_neighbor=pd.DataFrame(columns=['user_id','track_id','frequency','track','artist','TrackArtist'])
for item in neighbors:
    df_neighbor = df_neighbor.append(df_main[df_main.user_id == item],ignore_index = True)
df_neighbor
```

```python
df_neighbor = pd.DataFrame({'frequency':df_neighbor['frequency'], 'TrackArtist':df_neighbor['TrackArtist']})
```

```python
df_neighbor
```

```python
df_neighbor=df_neighbor.sort_values('frequency',ascending=False)
df_neighbor
```

```python
df_neighbor=df_neighbor.reset_index(drop=True)
df_neighbor
```
