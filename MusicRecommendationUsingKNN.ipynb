```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_main = pd.read_csv('SpotifyUpdated.csv')
df_main
```

```python
df_main["TrackArtist"] = df_main["track_name"] + " - " + df_main["artist_name"]
df_main
```

```python
len(df_main['TrackArtist'].unique()) # Unique users
```

```python
df_main['track_name'].value_counts() # Popular tracks
```

```python
len(df_main['User_ID'].unique()) # Unique users
```

```python
df_subset = df_main.drop_duplicates(['TrackArtist'])
df_subset = df_subset.reset_index(drop=True)
df_subset
```

```python
len(df_subset['track_id'].unique()) # Unique users
```

```python
import hashlib
df_subset['track_ids'] = [hashlib.blake2b(track.encode('utf-8')).hexdigest()[:24] for track in df_subset["TrackArtist"]]
df_subset
```

```python
len(df_subset['track_ids'].unique()) # Unique users
```

```python
df_subset.drop(['track_id'], axis = 1, inplace = True)
df_subset
```

```python
df_pivot=df_subset.pivot(index='User_ID',columns='track_ids',values='popularity')
df_pivot=df_pivot.fillna(0)
df_pivot
```

```python
df_pivot[df_pivot['000037d034fce952b54b7ecc']==1]
```

```python
from sklearn.neighbors import NearestNeighbors
knn=NearestNeighbors(n_neighbors=3,metric='cosine')
Model=knn.fit(df_pivot)
```

```python
df_pivot[df_pivot.index == 1]
```

```python
user = df_pivot.iloc[1,]
```

```python
distances, indices = Model.kneighbors([user])
distances, indices
```

```python
neighbors=[]
for item in indices[0][1:]:
    neighbors.append(df_subset[df_subset.index == item].User_ID.values[0])
neighbors
```

```python
df_neighbor=pd.DataFrame(columns=['User_ID','track_ids','Popularity','track_name','artist_name','TrackArtist'])
for item in neighbors:
    df_neighbor = df_neighbor.append(df_subset[df_subset.User_ID == item],ignore_index = True)
df_neighbor
```

```python
df_neighbor = pd.DataFrame({'popularity':df_neighbor['popularity'], 'TrackArtist':df_neighbor['TrackArtist']})
```

```python
df_neighbor
```

```python
df_neighbor=df_neighbor.sort_values('popularity',ascending=False)
df_neighbor
```

```python
df_neighbor=df_neighbor.reset_index(drop=True)
df_neighbor
```
